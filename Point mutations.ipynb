```python
ORIGINAL_SEQUENCE = 'TIIAL'
MUTATED_SEQUENCE = 'TIQAL'


Sf9_OptimalCodonTable = {
    'G': 'GGC', 'E': 'GAA', 'D': 'GAC', 'V': 'GTG', 'A': 'GCC',
    'R': 'AGA', 'S': 'TCG', 'K': 'AAA', 'N': 'AAC', 'M': 'ATG',
    'I': 'ATT', 'T': 'ACG', 'W': 'TGG', 'C': 'TGT', 'Y': 'TAC',
    'L': 'TTG', 'F': 'TTT', 'Q': 'CAA', 'H': 'CAC', 'p': 'CCG'
}

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio.Alphabet import IUPAC
from Bio.SeqFeature import SeqFeature


for record in SeqIO.parse("G:\\Andrii\\Constructs\\PSYR\\9792 Ct-4 9662-Rita M1-7946.gb", "genbank"):
    features = record.features
    sequence_bp = str(record.seq)
    sequence_aa = str(record.seq.translate())
    print(sequence_bp,'\n')
    print(sequence_aa,'\n')
    mut_position_aa = sequence_aa.find(ORIGINAL_SEQUENCE) + int(len(ORIGINAL_SEQUENCE)/2)
    print(mut_position_aa,'\n')
    mut_sequence_aa = sequence_aa.replace('TIIAL', 'TIQAL')
    print(mut_sequence_aa,'\n')
    mut_position_bp = mut_position_aa * 3
    mut_sequence_bp = sequence_bp[:mut_position_bp] + Sf9_OptimalCodonTable[MUTATED_SEQUENCE[2]] + sequence_bp[mut_position_bp+3:]
    print(Sf9_OptimalCodonTable[MUTATED_SEQUENCE[2]],'\n')
    print(mut_sequence_bp,'\n')
     
    
mut_sequence_bp_seqObject = Seq(mut_sequence_bp, IUPAC.unambiguous_dna)


recordToWrite = SeqRecord(mut_sequence_bp_seqObject,
                   name='Mutated_Sequence',
                   description='Full DNA sequence with a point mutation')
recordToWrite.features = features

output_file = open('G:\\Andrii\\Constructs\\PSYR\\mutationTest.gb', 'w')
SeqIO.write(recordToWrite, output_file, 'genbank')
output_file.close()
```
